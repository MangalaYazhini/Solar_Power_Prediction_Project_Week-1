#### Visualization

#### Univariate Analysis

```python
# Plot distribution of Generated Power
plt.figure(figsize = (6,6))
sns.histplot(df["generated_power_kw"] , kde = True)
plt.xlabel("Generated Power")
plt.ylabel("Frequency")
plt.title("Distribution of Generated Power")
plt.show()
```

#### Bivariate Analysis

```python
# Plot showing the relationship between the Angle of incidence and Generated power
plt.figure(figsize = (6,6))
sns.scatterplot(x = df["angle_of_incidence"], y = df["generated_power_kw"])
plt.xlabel("Angle of Incidence")
plt.ylabel("Generated Power")
plt.title("Relationship between the Angle of incidence and Generated power")
plt.show()
```
